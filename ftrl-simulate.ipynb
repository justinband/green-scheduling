```python
from mdps.latency import MDPLatency
from ftrl import FTRL
```

```python
NUM_STATES = 10
EPISODES = 30000
mdp = MDPLatency(NUM_STATES)
```

```python
ftl = FTRL(mdp, episodes=EPISODES)
ftl.reset()
```

```python
losses = ftl.execute()
```

```python
ftl.plot_loss(
    losses=losses,
    title=f"[FTRL] Total Energy Cost in {NUM_STATES}-State MDP with Latency",
    xaxis_title="# Episdoes",
    yaxis_title="Total Normalized Energy Cost"
    )
```
